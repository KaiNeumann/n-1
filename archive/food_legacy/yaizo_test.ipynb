```python
from Food import *

def parse_nutrients(nutrient_str):
    nutrient_dict = {}
    lines = nutrient_str.split('\n')
    skip = False
    for i in range(len(lines)-1):
        if skip:
            skip = False
            continue
        line = lines[i].strip()
        next_line = lines[i+1]

        if "<" in line or "kJ" in line or "pro Portion" in line or "Nährwerte" in line or "Vitamine" in line or "Mineralstoffe" in line:
            continue
        if "<" in next_line or "kJ" in next_line or "pro Portion" in next_line or "Nährwerte" in next_line or "Vitamine" in next_line or "Mineralstoffe" in next_line:
            continue

        nutriment = line.lower().replace('ungesät.','ungesättigte')
        value = 0

        if "kcal" in next_line:
            value = float(next_line.replace('kcal','').replace(',', '.'))
        elif "mg" in next_line:
            value = float(next_line.replace('mg','').replace(',', '.'))/1000
        elif "µg" in next_line:
            value = float(next_line.replace('µg','').replace(',', '.'))/1000000
        elif "g" in next_line:
            value = float(next_line.replace('g','').replace(',', '.'))
        
        if not value:
            continue

        nutrient_dict[nutriment] = value
        skip = True

    return nutrient_dict

nutrients = parse_nutrients(info)
ret = Food(nutrients)
print(nutrients)
print( f"Food({ret.nutrition_data})")
```
