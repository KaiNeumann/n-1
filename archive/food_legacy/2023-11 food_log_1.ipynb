```python
from functools import reduce

from Food import *
from food_naehrwertdaten_ch import *
from food_openfoodfacts_manual import *
from food_yazio_manual import *
from food_other_manual import *
from receipts import *
from health import *

################
# Tagebuch
################
diary = {}

diary["20231128"] = Day("20231128",
    haferflocken * gramm(75) + honig * gramm(5)
)


print("Combined Food:")
allFood = reduce(lambda x,y: x.add(y), [obj.food for obj in diary.values()], Food() )
print(f"= {allFood.weight/1000/len(diary)} kg per day, including beverages (water, beer, coffee,...)")
print(allFood)
print(allFood.compare_with_rdi("main",{},len(diary)))

print("-----")
comparisons = {datum: tag.calorie_balance() for datum, tag in diary.items()}
print(comparisons)
sum = reduce(lambda x,y: x + float(y.rstrip('%')) / 100, comparisons.values(), 0)
print(sum/len(diary))


balance = reduce(lambda sum, tag: sum + tag.food.nutrition_data.get('calories') * tag.food.weight / 100 - tag.activities.get_kcal(), diary.values(), 0)
print(f"{balance} kcal in {len(diary)} days")

print("-----")
print("Last day:")
latest = diary["20231128"]
print(latest.food)
print(latest.calorie_balance())
print(latest.compare_with_rdi("main"))
print(latest.compare_with_rdi("minerals"))
print(latest.compare_with_rdi("vitamins"))

```
