```python
info = """
Brennwert
108 kJ
Kalorien
26 kcal
Protein
1,3 g
Kohlenhydrate
3,2 g
davon Zucker
0 g
Fett
0,4 g
Ballaststoffe
0 g
Broteinheiten
0,3
Cholesterin
0 mg
"""
```

```python
from Food import *

def parse_nutrients(nutrient_str):
    nutrient_dict = {}
    lines = nutrient_str.split('\n')
    skip = True
    for i in range(len(lines)-1):
        if skip:
            #print("skip")
            skip = False
            continue
        line = lines[i].strip()
        next_line = lines[i+1]

        #print(f"{line} - {next_line}")

        if "Nährwerte" in line or "Vitamine" in line or "Mineralstoffe" in line:
            continue

        if "Nährwerte" in line or "Vitamine" in line or "Mineralstoffe" in line or "Brennwert" in line or "% Kohlenhydrate" in line or "% Fett" in line or "Broteinheiten" in line :
            skip = True
            continue
        if "kJ" in next_line or "Nährwerte" in next_line:
            skip = True
            continue

        nutriment = line.lower().replace('ungesät.','ungesättigte').replace("wassergehalt","water").replace("davon","")
        value = 0

        if "kcal" in next_line:
            value = float(next_line.replace('kcal','').replace(',', '.'))
        elif "mg" in next_line:
            value = float(next_line.replace('mg','').replace(',', '.'))/1000
        elif "μg" in next_line:
            value = float(next_line.replace('μg','').replace(',', '.'))/1000000
        elif "g" in next_line:
            value = float(next_line.replace('g','').replace(',', '.'))
        elif "%" in next_line:
            value = float(next_line.replace('%','').replace(',', '.'))
        
        if not value:
            continue

        #print(f"add {nutriment} = {value}")
        nutrient_dict[nutriment] = value
        skip = True

    return nutrient_dict

nutrients = parse_nutrients(info)
ret = Food(nutrients)

print( f"Food({ret.nutrition_data})")
```
